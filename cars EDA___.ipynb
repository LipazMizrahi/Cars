```python

df = pd.read_excel (r'cars_cleaning.xlsx', header=0)
df =  df[["price","area","producer","year","km","yad","gir","manoa_type","color","baalot","ac","windows","sunroof","gps","air_bag","numSTD"]]

df.columns
plt.hist(df["price"],bins=100)
df.info() 
```

# Categorical variables visualizations and analyzing




```python
A =  df[["price","area","producer","year","km","yad","gir","manoa_type","color","baalot","ac","windows","sunroof","gps","air_bag","numSTD"]]

corrMatrix = A.corr()

sns.heatmap(corrMatrix, 
          xticklabels=corrMatrix.columns.values,
           yticklabels=corrMatrix.columns.values,cmap='Blues',square=True)  

#print(corrMatrix)
#corrMatrix.to_excel("correlation.xlsx")
```
